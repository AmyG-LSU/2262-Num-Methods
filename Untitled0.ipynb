```python
def normalize_data(dataframe: pd.DataFrame):
    """
    Normalizes numerical columns in the DataFrame using min-max scaling.

    Parameters:
    - dataframe (pd.DataFrame): The DataFrame to normalize.

    Returns:
    - pd.DataFrame: A new DataFrame with normalized numerical columns.
    """
    df_normalized = dataframe.copy()
    numeric_columns = df_normalized.select_dtypes(
        include=['float64', 'int64']).columns

    for col in numeric_columns:
        min_col = df_normalized[col].min()
        max_col = df_normalized[col].max()
        df_normalized[col] = (df_normalized[col] -
                              min_col) / (max_col - min_col)

    return df_normalized
data = normalize_data(data)
X = data.drop('target', axis=1)
y = data['target']
print(data)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
log_reg = LogisticRegression()
log_reg.fit(X_train_scaled, y_train)
```

```python
y_test_pred = log_reg.predict(X_test_scaled)
print("Logistic Regression Test Accuracy:",
      accuracy_score(y_test, y_test_pred))
print(classification_report(y_test, y_test_pred))
```

```python

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

def crossValidate(data: pd.DataFrame, inputColumns: list, targetColumn: str, model: str, k_folds=10):
    """
    Performs k-fold cross-validation for the data.

    Parameters:
    - data: pd.DataFrame, the dataset.
    - inputColumns: list, the feature columns.
    - targetColumn: str, the target column.
    - model: str, the type of model to use ('logistic').
    - k_folds: int, number of folds (default 10).

    Returns:
    - float, the average accuracy across folds.
    """
    # Extract input features and target
    X = data[inputColumns]
    y = data[targetColumn]

    # Shuffle indices to randomize the dataset
    indices = np.arange(len(X))
    np.random.shuffle(indices)
    X = X.iloc[indices].reset_index(drop=True)
    y = y.iloc[indices].reset_index(drop=True)

    # Split into k folds
    folds = np.array_split(X, k_folds)
    folds_target = np.array_split(y, k_folds)

    scores = []  # List to store accuracy scores for each fold

    for i in range(k_folds):
        # Separate test fold
        test_fold = folds[i]
        test_target = folds_target[i]

        # Combine remaining folds for training
        train_folds = [folds[j] for j in range(k_folds) if j != i]
        train_targets = [folds_target[j] for j in range(k_folds) if j != i]
        train_fold = pd.concat(train_folds, axis=0)
        train_target = pd.concat(train_targets, axis=0)

        # Train and evaluate the logistic regression model
        if model == "logistic":
            clf = LogisticRegression(max_iter=1000)
            clf.fit(train_fold, train_target)
            predictions = clf.predict(test_fold)
            accuracy = accuracy_score(test_target, predictions)
            scores.append(accuracy)

    # Return the average accuracy across folds
    return np.mean(scores)
```
