# Q1 [10 marks] - Regular Expressions:
In this question you are to write a function that takes a string course code as input and returns some information about the course:
- Course department (we will only consider computer science and industrial engineering).
- Course level, i.e., 1000-level, 2000-level undergraduate course, etc. If a course number is 5000 or greater, it is just a graduate-level course.
- Whether the course has a separate section for undergraduate and graduate-level students. This is indicated by the suffix 'G' at the end of the course code.

So, in the case of a string like "CSC 4402G", you would return a succint response, i.e., "This is a 4000-level undergraduate computer science course. There is a section for graduate students."

Your code must be robust enough to handle different formats of course codes. Write it in the sample Python cell below and run it on the sample codes provided.

```python
## DO NOT MODIFY THESE
sample_course_codes = ["CSC2262", "csc 7300", "IE-3302", "IE7565", "CSC 4444G"]
## YOUR CODE BELOW HERE
import numpy as np
import re

def coursePrinter(x):

    textDept = "";
    textGraduate = "";
    textGraduateSection = "";

    text = x.replace(" ", "") #make easier to split

    text2 = re.split(r'\W?(\d+)', text, maxsplit = 1)
    #to split only dept and course details and g if there is one

    dept = text2[0]
    level = int(text2[1])

    # find if it has a graduate section
    textGraduateSection = "There is a section for graduate students" if text2[2] == "G" else ""

    #find level in text
    levelText = f"{(int(text2[1]) // 1000) * 1000}-level"

    #find the graduate level
    if level >= 5000:
        textGraduate = "graduate"
    else:
        textGraduate = "undergraduate"
    #find department name
    if dept == "CSC":
        textDept = "computer science"
    elif dept == "IE":
        textDept = "industrial engineering"

    #put it all together = )
    print("This is a " + levelText +" "+ textGraduate + " " + textDept + " course. " + textGraduateSection)


for x in sample_course_codes:
    coursePrinter(x)
```
